# Creating Plots using maltplotlib.pyplot

```python
import matplotlib.pyplot as plt
```

```python
x_axis = [1,2,3]
y_axis = [1,2,3]
plt.plot(x_axis, y_axis)
plt.show() # mainly use if your are working with .py file
```

### 1 liner plot of Sachin Tendulkar

```python
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
Tendulkar =  [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]
plt.plot(years, Tendulkar)
```

```python
# now let's make it more readable
plt.xlabel('Years')
plt.ylabel('Run scored')
plt.title('Sachin Tendulkar')
plt.plot(years, Tendulkar)
```

### 2 or more liner plot with proper labels

```python
kohli = [200,400,200,1400,800,400,1300,1100,900,700]
sharma = [700,300,500,200,800,1100,300,1000,600,800]

plt.plot(years, kohli, label='Virat Kohli')
plt.plot(years, sharma, label ='Rohit Sharma')

plt.xlabel('Years')
plt.ylabel('Run scored')
plt.title('Virat vs Rohit')
plt.legend()
# legend is better than list in legend because in larger data it can be prone to mismatch
plt.show()
```

### Styling

```python
# color, marker, line style and width
plt.plot(years, Tendulkar,'r>--', linewidth = 2,label = 'Sachin')
plt.plot(years, kohli,'go-.', linewidth = 2, label='Virat')
# so the following one is proper way to style a line in plot but if you don't want to write this much you can do the that way
plt.plot(years, sharma,color = 'blue',marker = 'x', linestyle = ':', linewidth = 2, label = 'Rohit')

# labeling 
plt.xlabel('Years')
plt.ylabel('Run scored')
plt.title('Logic less comparision for practice')
plt.legend() # we already saw this it is to show legends 

# grid 
plt.grid() # this will make a grid (for better understanding through naked eyes)
plt.tight_layout() # ig it won't show any difference but it is useful during .py file

# styling
plt.style.use('ggplot') # due to this style the grid is not visible, otherwise grid is there

plt.show()
```
