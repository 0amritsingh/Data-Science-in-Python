# Creating Histograms using Matplotlib


```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# the data
marks = np.array([
    6, 7, 5, 6, 7, 8, 7, 6, 8, 9,          
    8, 9, 10, 9, 10, 11, 10, 11, 12, 11,    
    12, 13, 12, 13, 13, 14, 13, 14, 15, 14,
    15, 16, 15, 16, 16, 17, 16, 17, 18, 17,
    18, 18, 19, 18, 19, 19, 20, 18, 19, 20,
    19, 20, 20
])
bins = [1,5,10,15,20]
mean = marks.mean()

# the histogram
plt.hist(marks, bins= bins, edgecolor = 'black')
plt.axvline(mean, linestyle = '--', color = 'red', linewidth = 1.25)

# labeling and styling
plt.grid(color = 'black', linestyle=':', alpha = 0.5)
plt.xlabel('Marks Range')
plt.ylabel('Monday per Week')
plt.legend([f'Avarage Marks ({round(mean, 2)})','Marks'], loc = 'upper left')
plt.title('Improvement in Marks in Monday Tests - 2024')

plt.show()
```
