# Creaking stack plots

```python
import matplotlib.pyplot as plt
```

We've seen pie chart which is good for fewer data insights but if we need a data that changes gradualy be time and we want insights over a time prieod then we use stacks. here is an example where we are observing acivities of a student after school in a week.

```python
# all data 
days = [1, 2, 3, 4, 5, 6, 7]  # Days of the week
studying = [3, 4, 3, 5, 4, 3, 4]
playing = [2, 2, 1, 1, 2, 3, 2]
watching_tv = [2, 1, 2, 2, 1, 1, 1]
sleeping = [5, 5, 6, 5, 6, 5, 5]

labels = ['Studying', 'Playing', 'Watching TV', 'Sleeping']
colors = ['red', 'skyblue', 'pink', 'grey']
plt.figure(figsize=(7,7)) # this is size of you graphs in inches

# graph
plt.stackplot(days, studying, playing, watching_tv, sleeping, labels=labels, colors=colors, alpha = 0.7)
# genral syntax of stackplot(x <what is on the x axix>, data of labels, labels, colors, alpha <opacity>)

# labeling and styling
plt.legend(loc='upper left')
plt.xlabel('Days')
plt.ylabel('Hours')
plt.title('Student Time Distribution')
plt.tight_layout()
plt.grid()
plt.style.use('classic')

plt.show()
```
