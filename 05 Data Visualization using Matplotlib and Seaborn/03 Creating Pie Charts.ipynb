# Creating Pie Chart


```python
import matplotlib.pyplot as plt

# you can pick a style initialy
plt.style.use('ggplot')
```

### normal pie chart

```python
players = ['Sachin', 'Virat', 'Rohit', 'KL Rahul']
runs = [1000,2300,1500,2000]
plt.pie(
    runs,
    labels=players
)
plt.title('Untitled Pie Chart')
plt.legend()
plt.tight_layout()
plt.show()
```

### styling and labeling and other addtional work

```python
# we can choose our custom colours
custom_colors = ["#C74545", "#56D36D", "#4A9DD4", "#DB4DBA"]
plt.pie(
    runs,
    labels=players,
    colors=custom_colors
)
plt.title('Untitled Pie Chart')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# adding edges or outline using wedgepros
plt.pie(
    runs,
    labels=players,
    colors=custom_colors,
    wedgeprops={'edgecolor': 'black', 'linewidth': 1.5, 'linestyle': '--'}
)
plt.title('Untitled Pie Chart')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# pull out a slice to highlight using explode
moveoutvirat = [0, 0.1,0,0,]
plt.pie(
    runs,
    labels=players,
    colors=custom_colors,
    explode=moveoutvirat
)
plt.title('Untitled Pie Chart')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# make it 3D using shadows
plt.pie(
    runs,
    labels=players,
    colors=custom_colors,
    explode=moveoutvirat,
    shadow=True # this will give the pie chart a 3D look
)
plt.title('Untitled Pie Chart')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# if you want to rotate you pie chart then use startangle
plt.pie(
    runs,
    labels=players,
    colors=custom_colors,
    explode=moveoutvirat,
    shadow=True,
    startangle=270
)
plt.title('Untitled Pie Chart')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# show percentage on each slice/partition of pie chart
plt.pie(
    runs,
    labels=players,
    colors=custom_colors,
    explode=moveoutvirat,
    shadow=True,
    startangle=270,
    autopct='%1.1f%%'
)
plt.title('Untitled Pie Chart')
plt.legend()
plt.tight_layout()
plt.show()
```
