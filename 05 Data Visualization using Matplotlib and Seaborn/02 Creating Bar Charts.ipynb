# Creating Bar Charts

```python
import matplotlib.pyplot as plt
import numpy as np
```

#### players data

```python
# All data we need
years = np.array([1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010])
sachin = np.array([500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500])
sehwag = np.array([80, 200, 900, 1400, 1600, 1800, 1500, 1100, 800, 200])
kohli  = np.array([100, 50, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100])
```

### Plotting a single bar graph

```python
plt.bar(years, sachin)
plt.xlabel('years')
plt.ylabel('run scored')
plt.title('Sachin Tendulkar')
plt.show()
```

```python
# we can also change bar width and colour and over all style of the chart
plt.bar(years, sachin, width=1, color='red', label = 'Sachin')
plt.xlabel('years')
plt.ylabel('run scored')
plt.title('Sachin Tendulkar')
plt.tight_layout()
plt.legend()
plt.style.use('classic')
plt.show()
```

### Plotting a bar graph of multiple bars

```python
x = np.arange(len(years))
width = 0.25

# bar graph
plt.bar(x - width, sachin, color='red',width=width, label='Sachin')
plt.bar(x, kohli, color='green',width=width, label='Virat')
plt.bar(x + width, sehwag, color='blue',width=width, label='Sehwag')

# xticks
plt.xticks(x,years)

# styling and labeling
plt.xlabel('Years')
plt.ylabel('Run Scored')
plt.title('Sachin vs Kohli vs Shewag')
plt.legend()
plt.tight_layout()
plt.style.use('classic')

plt.show()
```

### horizontal bar graph

```python
# data
players = ['Sachin', 'Kohli', 'Sehwag']
run_10yrs = [sachin.sum(), kohli.sum(), sehwag.sum()]

# horizontal graph
plt.barh(players, run_10yrs, color = 'skyblue')

# labeling
plt.xlabel('Run Scored in 10 years (1990-2010)')
plt.ylabel('Players')
plt.title('Nonsense comparrision')
plt.tight_layout()

plt.show()
```

### writing above the bars of graph values using `plt.text()`

```python
# data
players = ['Kohli', 'Sachin', 'Sehwag']
runs = [ kohli.sum(), sachin.sum(),sehwag.sum()]

# graph
plt.bar(players, runs, color = "#b58a3f")

# values above each bar
for i in range(len(runs)):
    plt.text(i, runs[i] + 250, str(runs[i]), ha='center')

# labeling and styling
plt.xlabel('Players')
plt.ylabel('Score')
plt.title('Untitled')
plt.tight_layout()
plt.style.use('dark_background')

plt.show()
```
