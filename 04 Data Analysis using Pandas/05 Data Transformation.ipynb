## Reindexing

```python
df2 = df.sort_values('year').copy() # making new dataset
df2
```

```python
# as you can see the index is not ordered
df2.reset_index() 
# but this will give you a new column of old index
```

```python
# but we don't want that old index so 
df2.reset_index(drop=True)
```

```python
# but as we new these all changes are just a preview
df2
# see no change
```

```python
# if we want make them permanenet or direcly in the dataframe
# we need to use inplace arguement
df2.reset_index(drop=True,inplace=True)
```

**NOTE:** `inplace=True` is an arguement use with most of the pandas's methods.
It makes the changes directly in the dataframe so we don't need to make copy of it.

```python
df2 # as you can see its changes in dataframe
```

## Rank

```python
df2['rank'] = df2['imdb'].rank(ascending=False) 
df2
```

```python
# as you can see at index 0 and 4 have same imdb so the rank is in average
# to avoid that we can use method argument
df2['rank'] = df2['imdb'].rank(ascending=False, method='dense') 
df2
```

## Rename columns

```python
df2.rename(columns={"film": "movies"}, inplace=True)
# also can change index
df2.rename(index={0: "row1", 1: "row2"}, inplace=True)
df2
```

```python
# rename all columns and index
df2.columns = ['actor', 'film', 'year', 'genre','boxoffice_usd', 'imdb', 'rank']
df2.index = [i for i in range(6)]
df2
```

```python
# you can also change order of columns
df2 = df2[['actor','boxoffice_usd', 'year', 'genre', 'imdb', 'rank', 'film']]
df2
```

```python
# we can also move one column to front by using python logic
cols = ['genre'] + [col for col in df2.columns if col != 'genre']
df2 = df2[cols]
df2
```
