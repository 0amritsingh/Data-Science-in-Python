# Data Frames

````python
import pandas as pd
import numpy as np
````

Python list

````python
arr1 = [['Amrit', 18], ['John', 23], ['Mebal', 12], ['Dipper', 12], ['Stan', 76], ['Kenney', 17]]
df = pd.DataFrame(arr1, columns=['Name', 'Age'])
df
````

Python Dictionary

````python
arr2 = {
    'Name': ['Eran', 'Mikasa', 'Armin', 'Jeon', 'Sasha', 'Connie'],
    'Age': [52,61,42,32,41,32]
}
df = pd.DataFrame(arr2)
df
````

NumPy Array

````python
arr3 = np.array([['Puma', 1990], ['Adidas', 1870], ['Nike', 1997]])
df = pd.DataFrame(arr3, columns=['Brand', 'Since'])
df
````

Excel Spreadsheet

````python
df = pd.read_excel('data.xlsx')
df
````

````python
df = pd.read_csv('data.csv')
df
````

## EDA (Exploratory Data Analysis)
Exploratory Data Analysis (EDA) is an essential first step in any data science project.
It involves taking a deep look at the dataset to understand its structure, spot patterns, identify anomalies, and uncover relationships between variables.
```python
df.head()         # First 5 rows
df.tail()         # Last 5 rows
df.info()         # Column info: types, non-nulls
df.describe()     # Stats for numeric columns
df.columns        # List of column names
df.shape          # (rows, columns)
```
