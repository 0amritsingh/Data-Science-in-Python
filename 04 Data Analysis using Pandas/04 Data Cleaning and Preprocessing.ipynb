```python
import pandas as pd
```

## Null (NaN) values

```python
# Data
df = pd.read_csv('clean_the_data.csv')
df
```

```python
# Identify the null (NaN) values in the data
df.isnull()
```

```python
# Total null values in each coloumn (default axis)
df.isnull().sum()
```

```python
# Total null values in each rows (axis = 1)
df.isnull().sum(axis=1)
```

```python
# remove all rows have null value
df.dropna()
```

```python
# remove all columns have null value
df.dropna(axis=1)
```

```python
# Filling the null values
df.fillna('something')
```

```python
# you can also fill in spcific column
df['Age'].fillna(df['Age'].mean())
```

```python
# forward fill - fill same as the the value above the null value
df.ffill()
```

```python
# backward fill - fill same as the the value below the null value
df.bfill()
```

## Duplicated values

```python
# Identifing duplicate values
df.duplicated()
```

```python
# Identifing duplicate values in spicific column
df.duplicated(subset=['Name'])
```

```python
# Removing duplicate rows
df.drop_duplicates()
```

```python
# Remove by combination, here if someone's name and his department are duplicate then it will be removed
df.drop_duplicates(subset=['Name', 'Department'])
```

## Changing type

```python
df.info() # as you can see age is float, we want to change its type 
```

```python
df['Age'].astype('int32')
# as you can see here is an error
# becuase our dataset has NaN values also
# we will make a copy of df and with no null values
```

```python
df2 = df.ffill().copy()
df2
```

```python
df2['Age'] = df2['Age'].astype('int32')
df2.info() # as you can see Age is now int
```

## Use `apply()`, `map()`, `replace()` to transform your columns

```python
df3 = df.ffill().copy() # our dataset
df3
```

Apply

```python
df3['Heirarchy'] = df3['Age'].apply(lambda x: 'Senior' if x > 30 else 'Junior')
df3
```

Map

```python
new_gender = {'M':'Male', 'F':'Female', 'O':'Other'}
df3['Gender'] = df3['Gender'].map(new_gender)
df3
```

Replace

```python
new_department = {'IT': 'Information Technology', 'HR': 'Human Resources'}
df3['Department'] = df3['Department'].replace(new_department)
df3
```
