```python
df.at[8, 'Rating'] # At is based on label
```

```python
df.iat[8, 7] # iAt based indexing can only have integer indexers
```

```python
print(df.loc[8, 'Rating'])
print(df.iloc[8,7])
# loc(label based) and iloc(index based) can do the same thing as iat but - It is made of SLICING.
```

```python
df.loc[0:3, ['Name', 'Age', 'Salary']] # we gave 0:3 and we get 0,1,2,3 and given list of columns
```

```python
df.iloc[0:3, 1:4] # we gave 0:3 but it gives only 0,1,2 but not 3 also in columns 1:4 we get only three
```

```python
df[df['Age'] > 40] # we can give condition
```

```python
df[df['Age'] == 41]
```

```python
df[(df['Age'] > 20) & (df['Age'] < 50)] # & [and percent] means 'AND'
```

```python
df[(df['Age'] > 50) | (df['Rating'] > 4)] # | (pipe) means 'OR'
```

```python
df.query("Department == 'IT' and Rating > 3.5") # you can use 'and', 'or', 'not instead of &, |, ~
```

```python
df.query("`Is Permanent` == True") # if there is space in column's name then you backticks (``)
```
