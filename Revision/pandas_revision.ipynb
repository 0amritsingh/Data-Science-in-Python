```python
data = {
    'Name':['Amrit', 'Manav', 'Vivek'],
    'Maths':[89,33,78],
    'Physics':[78,93,29],
    'Chemistry':[48,39,23]
}
marks = pd.DataFrame(data)
marks
```

```python
melted_marks = marks.melt(id_vars='Name', value_vars=['Maths', 'Physics', 'Chemistry'], var_name='Subjects', value_name='Marks')
melted_marks
```

```python
pivoted_marks = melted_marks.pivot(index='Name', columns='Subjects', values='Marks')
pivoted_marks
```

```python
# but pivot can throw errors if the vlaues are duplicate like
data2 = {
    'Name':['Amrit', 'Amrit', 'Manav', 'Manav'],
    'Stream':['PCM','PCM','PCM','PCM'],
    'Percentage':[80,70,90,58]
}
marks2 = pd.DataFrame(data2)
marks2
```

```python
marks2.pivot(index='Name', columns='Stream', values='Percentage')
```

```python
# as you can see (read the last line for the error), so to prevent this we need to use pivot_table.
marks2.pivot_table(index='Name', columns='Stream', values='Percentage', aggfunc='min')
# its requares a aggregate function, it can be anything like mean, count, min, max, etc.
```

```python
corporate = pd.DataFrame({
    "Department": ["HR", "HR", "IT", "IT", "Marketing", "Marketing", "Sales", "Sales"],
    "Team": ["A", "A", "B", "B", "C", "C", "D", "D"],
    "Gender": ["M", "F", "M", "F", "M", "F", "M", "F"],
    "Salary": [85, 90, 78, 85, 92, 88, 75, 80],
    "Age": [23, 25, 30, 22, 28, 26, 21, 27],
    "JoinDate": pd.to_datetime([
        "2020-01-10", "2020-02-15", "2021-03-20", "2021-04-10",
        "2020-05-30", "2020-06-25", "2021-07-15", "2021-08-01"
    ])
})  

corporate
```

```python
corporate.groupby(['Team'])['Salary'].agg('mean')
```

```python
corporate.groupby(['Department', 'Gender'])['Salary'].agg(['min', 'mean', 'max'])
```

```python
corporate['Ava_Salary'] = corporate.groupby('Department')['Salary'].transform('mean')
corporate
```
